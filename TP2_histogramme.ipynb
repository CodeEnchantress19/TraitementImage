Exercices 1 : histogramme 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('fichier.png', cv2.IMREAD_GRAYSCALE)

histogram = np.zeros(256, dtype=int)


for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel_value = image[i, j]
        histogram[pixel_value] += 1

plt.bar(np.arange(256), histogram)
plt.xlabel('Niveau de gris')
plt.ylabel('Fréquence')
plt.show()
```

Exercice 2 : histogramme cumulé

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('fichier.png', cv2.IMREAD_GRAYSCALE)

rows, cols = image.shape

cumulative_histogram = np.zeros(256, dtype=int)

for pixel_value in range(256):
    count = 0  
    
    
    for i in range(rows):
        for j in range(cols):
            if image[i, j] == pixel_value:
                count += 1
    
    if pixel_value == 0:
        cumulative_histogram[pixel_value] = count
    else:
        cumulative_histogram[pixel_value] = cumulative_histogram[pixel_value - 1] + count

plt.plot(cumulative_histogram)
plt.xlabel('Niveau de gris')
plt.ylabel('Histogramme cumulé')
plt.show()
```

Exercice 3 : normarliser les valeurs de l’histogramme cumulé

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Chargez l'image en niveaux de gris
image = cv2.imread('fichier.png', cv2.IMREAD_GRAYSCALE)

histogram = np.zeros(256, dtype=int)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel_value = image[i, j]
        histogram[pixel_value] += 1


cumulative_histogram = np.cumsum(histogram)

Min = np.min(cumulative_histogram)
Max = np.max(cumulative_histogram)

normalized_cumulative_histogram = np.zeros_like(cumulative_histogram, dtype=float)
for i in range(256):
    normalized_cumulative_histogram[i] = (cumulative_histogram[i] - Min) / (Max - Min) * 255

plt.plot(normalized_cumulative_histogram)
plt.xlabel('Niveau de gris')
plt.ylabel('Histogramme cumulé normalisé')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Chargez l'image en niveaux de gris
image = cv2.imread('fichier.png', cv2.IMREAD_GRAYSCALE)

# Affichez l'image initiale
plt.imshow(image, cmap='gray')
plt.title('Image avant conversion')
plt.axis('off')  # Pour supprimer les axes
plt.show()

# Calculez l'histogramme
histogram = np.zeros(256, dtype=int)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel_value = image[i, j]
        histogram[pixel_value] += 1



# Trouvez Imin et Imax
Imin = np.min(image)
Imax = np.max(image)

# Appliquez la conversion aux niveaux d'image
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel_value = image[i, j]
        if pixel_value <= 120:
            image[i, j] = 0
        elif pixel_value >= 200:
            image[i, j] = 255
        else:
            image[i, j] = int((pixel_value - Imin) / (Imax - Imin) * 255)

# Affichez l'image résultante après la conversion
plt.figure()
plt.imshow(image, cmap='gray')
plt.title('Image après conversion')
plt.axis('off')  # Pour supprimer les axes
plt.show()
```
