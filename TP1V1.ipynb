```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger l'image Lenna
image = cv2.imread('Lenna.png')

# Séparer l'image en quatre quadrants
hauteur, largeur, _ = image.shape
moitie_hauteur = hauteur // 2
moitie_largeur = largeur // 2

quadrant_sup_gauche = image[:moitie_hauteur, :moitie_largeur]
quadrant_sup_droit = image[:moitie_hauteur, moitie_largeur:]
quadrant_inf_gauche = image[moitie_hauteur:, :moitie_largeur]

# Supprimer les composantes couleurs des quadrants selon les spécifications
quadrant_sup_gauche[:, :, 0] = 0  # Effacer la première composante (bleu)
quadrant_sup_droit[:, :, 1] = 0  # Effacer la deuxième composante (vert)
quadrant_inf_gauche[:, :, 2] = 0  # Effacer la troisième composante (rouge)

# Réunir les quadrants pour former l'image résultante
image_modifiee = np.vstack((np.hstack((quadrant_sup_gauche, quadrant_sup_droit)),
                            np.hstack((quadrant_inf_gauche, image[moitie_hauteur:, moitie_largeur:]))))

# Afficher l'image résultante
plt.imshow(cv2.cvtColor(image_modifiee, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Pour supprimer les axes
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger l'image Lenna
image = cv2.imread('Lenna.png')

# Séparer les composantes de couleur (rouge, vert, bleu)
rouge = image.copy()
rouge[:, :, 1] = 0  # Mettre à zéro la composante verte
rouge[:, :, 2] = 0  # Mettre à zéro la composante bleue

vert = image.copy()
vert[:, :, 0] = 0  # Mettre à zéro la composante rouge
vert[:, :, 2] = 0  # Mettre à zéro la composante bleue

bleu = image.copy()
bleu[:, :, 0] = 0  # Mettre à zéro la composante rouge
bleu[:, :, 1] = 0  # Mettre à zéro la composante verte

# Afficher les trois images séparées
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(cv2.cvtColor(rouge, cv2.COLOR_BGR2RGB))
axes[0].set_title('Composante Rouge')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(vert, cv2.COLOR_BGR2RGB))
axes[1].set_title('Composante Verte')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(bleu, cv2.COLOR_BGR2RGB))
axes[2].set_title('Composante Bleue')
axes[2].axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger l'image Lenna
image = cv2.imread('Lenna.png')
# Transformer l'image en niveaux de gris en appliquant la formule
niveaux_de_gris = np.zeros((image.shape[0], image.shape[1]))

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        b, g, r = image[i, j]
        niveau_gris = int(0.2126 * r + 0.7152 * g + 0.0722 * b)
        niveaux_de_gris[i, j] = niveau_gris

# Afficher l'image en niveaux de gris

plt.imshow(niveaux_de_gris, cmap='gray')
plt.show()
```
